```python
import numpy as np
import matplotlib.pyplot as plt
```

# 그래프 주석 달기

* plt.title() : 캔버스의 타이틀을 표시 (한글은 잘 안됨)
* Axes.set_title() : 서브 플롯에 타이틀 표시

```python
fig = plt.figure()
x = np.linspace(0,12,60)
y = np.sin(x)
plt.plot(x,y)
plt.title('Title!')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
x = np.linspace(0,12,60)
y = np.sin(x)
y2 = np.cos(x)

ax1.plot(x,y)
ax1.set_title('sub plot1')
ax2.plot(x,y2)
ax2.set_title('sub plot2')
plt.show()
```

# 그래프 범례 추가하기

* .legend() : plot의 label에 따른 범례를 표에 표시한다.

```python
fig = plt.figure()
x = np.linspace(0,12,60)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y, label='label1')
plt.plot(x,y2, label='label2')
plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
x = np.linspace(0,12,60)
y = np.sin(x)
y2 = np.cos(x)

ax1.plot(x,y, label='label1')
ax1.set_title('sub plot1')
ax1.legend()
ax2.plot(x,y2, label='label2')
ax2.set_title('sub plot2')
ax2.legend()
plt.show()
```

# 좌표축 커스텀

* plt.xlim(), plt.ylim() : 축의 시작점, 종료점을 표시함
* plt.xticks(), plt.yticks() : 해당 축만 표시한다
    * xticks로 텍스트도 들어갈 수 있음
* plt.axis() : x,y,에 대한 min, max값을 지정할 수 있음

* Axes.set_xlim(), Axes.set_ylim() : 같은 역할
* Axes.set_xticks(), Axes.set_yticks() : 같은 역할



```python
fig = plt.figure()
x = np.array([1,2,3,4])
y = np.array([10, 9.2, 10.5, 9])
plt.xlim(0.5, 4.5)
plt.ylim(3, 11)
# plt.ylim(8, 11.5)
plt.plot(x,y)

plt.show()
```

```python
fig = plt.figure()
x = np.array([1,2,3,4])
y = np.array([10, 9.2, 10.5, 9])
plt.xticks(np.array([2.5, 3.0, 3.5, 3.7, 4.0]))
plt.yticks(y)
plt.plot(x,y)

plt.show()
```

```python
fig = plt.figure()
x = np.array([1,2,3,4])
y = np.array([10, 9.2, 10.5, 9])

plt.plot(x,y)
plt.axis([1,5,7, 11])
plt.show()
```

```python
fig = plt.figure()
x = np.array([1,2,3,4])
y = np.array([10, 9.2, 10.5, 9])
plt.xticks(x, ['one', 'two', 'three', 'four!'])
plt.yticks(y)
plt.plot(x,y)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
x = np.linspace(0,12,60)
y = np.sin(x)
y2 = np.cos(x)

ax1.plot(x,y, label='label1')
ax1.set_title('sub plot1')
ax1.legend()
ax1.set_xlim(0.4, 8.5)
ax1.set_ylim(-1, 4)
ax2.plot(x,y2, label='label2')
ax2.set_title('sub plot2')
ax2.legend()
ax2.set_xticks(np.array([2,3,4,7,8]))
ax2.set_yticks(np.array([-1, -0.1, 0.3, 0.8, 1.0]))
ax2.axis([1,5,-2,4])
plt.show()
```

#  함수를 위한 그래프 그리기

그래프의 중심을 이동하려면 좌표축에 해당하는 spines를 이동해야함.

```python
ax = plt.gca()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

x = np.arange(-10, 13)
y = 3*x

plt.plot(x,y)
plt.show()
```
