# 캔버스와Axes

* plt.figure() : 캔버스 생성
    * fig.add_subplot(r,c,n) : 또다른 Axes를 추가한다. r * c의 서브 플롯을 생성하고 n번째 서브 플롯에 위치시킨다.
* plt.axes() : 객체 만들기

* plt.subplots() : 캔버스와 객체를 동시에 만든다
    * Axes.plot() : 그래프를 그린다.
        * 굳이 Axes객체에 하지 않고 plt.plot()으로 그려도 자동으로 Axes 객체가 만들어짐.
        
    * nrow, ncol을 지정하여 여러개의 Axes 객체를 만들 수 있다.
    
* .plot() : fig, 또는 axes에서 그래프를 그린다.
    * ls : 라인스타일 지정
        * '-'
        * '--'
        * '-.'
        * ':'
    * color : 색 지정
        * 0~1실수 : 흑백
        * hexcode : #00ff00
        * RGB or RGBA : (0.5, 0.3, 0.3. 0.1)        
    * marker : 마커 지정 (입력된 x,y의 지점마다 생긴다.)
        * '+'
        * 'o'
        * '*'
        * 's'
        * '.'
    * 문자열 처리 : 위 파라미터를 하나의 문자열로 인식하여 표시함
        * ex) r- : 빨간색 실선
        * ex) bs : 파란색 네모
        * ex) g^ : 초록색 세모

```python
fig = plt.figure(figsize=(8,6))
plt.axes()
print(fig)

plt.show()
```

# 그래프그리기

## plt.plot()

```python
fig = plt.figure()
x = np.linspace(0,12,60)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

# ax.plot()

```python
fig, ax = plt.subplots()
x = np.linspace(0,12,60)
y = np.sin(x)
ax.plot(x,y)
plt.show()
```

# 두개의그래프

```python
fig = plt.figure()
x = np.linspace(0,12,60)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y)
plt.plot(x,y2)
plt.show()
```

# 두개의Axes객체

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
x = np.linspace(0,12,60)
y = np.sin(x)
y2 = np.cos(x)

ax1.plot(x,y)
ax2.plot(x,y2)
plt.show()
```

# 여러개의Axes객체

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(2,4,1)
ax2 = fig.add_subplot(2,4,7)
ax1.plot(x,y)
ax2.plot(x,y2)
plt.show()
```

# 다양한모양

```python
fig = plt.figure(figsize=(10,5))
x = np.linspace(0,12,60)
y = np.sin(x)

ax1 = fig.add_subplot(1,4,1)
ax2 = fig.add_subplot(1,4,2)
ax3 = fig.add_subplot(1,4,3)
ax4 = fig.add_subplot(1,4,4)

ax1.plot(x,y, ls='-', color='#ff0000')
ax2.plot(x,y, ls='--', color = '0.4')
ax3.plot(x,y, ls='-.', color = (0.3, 0.5, 0.9))
ax4.plot(x,y, ls=':')
plt.show()
```

```python
fig = plt.figure()
x = np.linspace(1, 10, 30)
plt.plot(x, x, 'r-', x, x**2, 'bs', x, x**3, 'g^')
plt.show()
```
